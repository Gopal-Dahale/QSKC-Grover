### Multiplier

```python
qc = QuantumCircuit(12,4)
qubits = [
    [0,6,11],
    [1,5,11],
    [2,4,11],
    [0,5,10],
    [1,4,10],
    [0,4,9],
    [9,8],
    [10,9],
    [11,10],
    [0,7,8],
    [1,6,8],
    [2,5,8],
    [3,4,8],
    [1,7,9],
    [2,6,9],
    [3,5,9],
    [2,7,10],
    [3,6,10],
    [3,7,11]
]


for i in qubits:
    if len(i) == 3:
        qc.toffoli(i[0],i[1],i[2])
    else:
        qc.cnot(i[0],i[1])

qc.measure([8,9,10,11],[0,1,2,3])

qc.draw()
```
